# Data Quality Report

## 1. Case Mismatch in Type Column

The **Type** column contains inconsistent capitalization. The same category appears in multiple formats such as:

* Movie
* movie
* Series
* series
* Documentary
* documentary

These values represent the same content types and should be standardized using `.str.title()`.

---

## 2. Trailing Spaces in Genre Column

The **Genre** column contains trailing spaces in several categories. Examples include:

* Action and Action
* Drama and Drama
* Comedy and Comedy
* Romance and Romance
* Horror and Horror
* Thriller and Thriller
* Kids and Kids
* Sci-Fi and Sci-Fi
* Reality and Reality

These values are treated as different categories by Python and should be cleaned using `.str.strip()`.

---

## 3. Trailing Spaces in Language Column

The **Language** column also contains trailing spaces. Examples include:

* Hindi and Hindi
* English and English
* Tamil and Tamil
* Telugu and Telugu
* Turkish and Turkish
* Kannada and Kannada

These values should be standardized using `.str.strip()`.

---

## 4. Missing Director Values

The dataset contains blank (missing) values in the **Director** column. These missing values should be identified and handled appropriately during cleaning.

---

## 5. Missing AddedDate Values

Some records have missing values in the **AddedDate** column. Since this field is required for time-based analysis, it should be converted to datetime format using:

`pd.to_datetime(errors='coerce')`

---

## 6. Out-of-Range IMDb Scores

The valid range for **IMDbScore** is 1 to 10. Any value below 1 or above 10 is considered invalid data and should be replaced with NaN.

---

## 7. Out-of-Range Critic Ratings

The valid range for **CriticRating** is also 1 to 10. Values outside this range indicate data quality issues and should be replaced with NaN.

---

## 8. Absurd Runtime Values

The **RuntimeMinutes** column contains unrealistic runtime values (extreme outliers). Such values are likely data-entry errors and should be replaced with NaN.

---

## 9. Negative Subscriber Counts

The **SubscribersGainedThousands** column contains negative values. Since subscriber gains cannot be negative, these values should be replaced with 0.

---

## 10. Duplicate Rows

The dataset contains duplicate records. Duplicate rows should be removed before performing any analysis to ensure accurate results.


```python
text_cols = [
    'Genre',
    'Language',
    'Country',
    'Director',
    'AgeRating',
    'TrendingStatus'
]

for col in text_cols:
    df[col] = df[col].str.strip()
```

```python
df['Type'] = df['Type'].str.title()
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
df.loc[
    (df['IMDbScore'] < 1) |
    (df['IMDbScore'] > 10),
    'IMDbScore'
] = np.nan
```

```python
df.loc[
    df['RuntimeMinutes'] > 1000,
    'RuntimeMinutes'
] = np.nan
```

```python
df.loc[
    df['SubscribersGainedThousands'] < 0,
    'SubscribersGainedThousands'
] = 0
```

```python
df['AddedDate'] = pd.to_datetime(
    df['AddedDate'],
    errors='coerce'
)
```

```python
df['Profit_Cr'] = (
    df['RevenueCr']
    - df['ProductionCostCr']
)
```

```python
df['ROI_Pct'] = (
    df['Profit_Cr']
    / df['ProductionCostCr']
) * 100
```

```python
df['Performance_Band'] = np.where(
    df['Profit_Cr'] > 20,
    'Hit',
    np.where(
        df['Profit_Cr'] >= 0,
        'Break-even',
        'Flop'
    )
)
```

```python
df.to_csv('../output/cleaned_cinestream.csv',index=False)
```
